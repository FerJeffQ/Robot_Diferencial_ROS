### Instalacion de ROSSERIAL
Instructivo para instalar puerto serial con ros en arduino: 
http://wiki.ros.org/rosserial_arduino/Tutorials/Arduino%20IDE%20Setup

``` sudo apt-get install ros-noetic-rosserial ```

### Otorgar permisos a Puerto arduino

``` sudo chmod 777 /dev/ttyACM0 ```

## Configuracion RED
### MASTER PC
``` echo 'export ROS_IP="192.168.1.30"' >> ~/.bashrc ```

``` echo 'export ROS_MASTER_URI=http://"192.168.1.30:11311"' >> ~/.bashrc ```

``` source ~/.bashrc ```

### SCLAVE RASPBERRY

``` echo 'export ROS_IP="192.168.1.41"' >> ~/.bashrc```

``` echo 'export ROS_MASTER_URI=http://"192.168.1.30:11311"' >> ~/.bashrc```

``` source ~/.bashrc ```

## Para configuracion de paquete de ROS:
Creamos la carpeta con 

``` mkdir Entorno_robot/src```

``` catkin_make ```

Agregamos la direccion nuestro entorno al de bash:

``` echo " source /home/ferjeff/Entorno_robot/devel/setup.bash" >> ~/.bashrc ```

``` source ~/.bashrc ```

Dentro de la carpeta /src creamos el paquete ROS con las dependencias:

```catkin_create_pkg robot_diff std_msgs rospy roscpp urdf xacro move_base_msgs nav_msgs robot_pose_ekf robot_state_publisher tf ```

## Para ejecutar el Robot diferencial se utiliza 

### En PC
``` roslaunch robot_diff robot_diff_arduino.launch```

## En raspberry
``` ssh ubuntu@192.168.1.41```
 
``` rosrun rosserial_python serial_node.py /dev/ttyACM0 ```
